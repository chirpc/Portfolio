## FEATURE ENGINEERING

We would like to add additional features to our dataset to ultimately improve our model.

This process (which I perform in a seperate notebook) includes:

1. Extracting the respondents region based on approximate longitude, latitude coordinates using argis reverse geocoding method

2. Computing the number of financial institutions or access points within a region

3. Computing the nearest distance of respondent from financial institution or access points using k-nearest neighbour algorithm and haversine distance formula.

-------

The idea here is that the closer a respondent is to financial access points, the higher the probability of adoption/take-up. We also want to control for the composition of this data points in a specific area (region/ward). This feature should tell us if having more access points within an area would increas take-up. This would be somewhat analogous to having data on the population of an area, then we could account for the number of financial access points per capita. Instead we can measure the average distance from financial access point per person.

- How does financial access compare across regions?
- Does proximity affect adoption?

For example, if an area has less access points, that could explain why take-up would also be low. Im that case, business might want to increase their accessibility to increase take-up. Aggregate Financial access points by ward to determine area composition
