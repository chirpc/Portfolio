```python
##LIBRARIES##
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
import random
import math

#ML
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder, KBinsDiscretizer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC, NuSVC
from xgboost import XGBClassifier

#SPATIAL
import geoviews as gv
import geoviews.feature as gf
import xarray as xr
from cartopy import crs
from geoviews import dim, opts
gv.extension('bokeh')
import geopandas as gpd
import geoviews.tile_sources as gts


##FUNCTIONS##
def data_pipeline(data, output):
    """
    Preprocessing pipeline part 1: Transform full data frame
    Arguments: Pandas dataframe, output column (dependent variable)
    Returns: Modified dataframe
    """
    data = cost_per_metric(data, output) if 'cost_per' in output \
                                         else data[data[output] > 0]
    data = drop_columns_missing(data, output, threshold=.5)
    #data = data.dropna(axis='index')
    return data


def cost_per_metric(data, output):
    """Create 'cost_per_...' column and remove data where output is 0 or NaN"""
    metric = output.replace('cost_per_', '') + 's'
    data = data[data[metric] > 0]
    data.insert(0, output, [row['cost'] / row[metric]
                            for index, row in data.iterrows()])
    return data


def drop_columns_missing(data, output, threshold=0.5):
    """Drop columns with more than threshold missing data"""
    #recode "NA" and "don't know" values to missing
    data = data.replace(-1,np.nan)
    #checking the number of missing data in columns that contain missing data
    null_df = pd.DataFrame(data.isnull().sum()/len(data)*100)
    print(null_df[null_df[0]!=0])
    rows = data[output].count()
    for column in list(data.columns):
        if data[column].count() < rows * threshold:
            data = data.drop([column], axis=1)
    return data


def drop_columns_outlier(data, output, threshold=0.95):
    """Drop columns with more than threshold outlier data"""
    for column in data.columns.difference(['output']): 
        if data[column].value_counts(normalize=True).max(axis='rows') > 0.95:
            data.drop(columns=column,inplace=True)
    return data

        
def create_other_buckets(data, features, threshold=0.1):
    """Put rare categorical values into other bucket"""
    for column in features:
        results = data[column].count()
        groups = data.groupby([column])[column].count()
        data["{}_x".format(column)] = data[column]
        for bucket in groups.index:
            if groups.loc[bucket] < results * threshold:
                data.loc[data["{}_x".format(column)] == bucket, "{}_x".format(column)] = "other"
        data.drop(columns=[column], axis=1, inplace=True)
    return data

def one_hot_encode(data, features):
    """One-hot encode categorical data"""
    #categoricals = list(data.select_dtypes(include='object').columns) 
    data = data[features].apply(lambda x: x.astype(str))
    ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
    train_ohe = ohe.fit_transform(data[features])
    data_ohe = pd.DataFrame(train_ohe, columns=ohe.get_feature_names_out())
    data_ohe = data_ohe.apply(lambda x: x.astype(int))
    for column in features:
        if 'other' in data[column].unique():
            data_ohe = data_ohe.drop([column + '_other'], axis=1)
        else:
            excl_val = list(data[column].sort_values(na_position='last', ascending=False).unique())[0]
            data_ohe = data_ohe.drop([column + '_{}'.format(excl_val)], axis=1)   
        try:
            if 'nan' in data[column].unique():
                data_ohe = data_ohe.drop([column + '_nan'], axis=1) 
        except Exception as e:
            print(e)
    return data_ohe 
       
        
def one_hot_encode_binary(data, features):
    """One-hot encode binary categorical data"""
    for column in features:
        data["{}_1".format(column)] = [1 if x==1 else 0 for x in data[column]]
        data.drop(columns=[column], axis=1, inplace=True)
    return data
    
'''
    for column in features:
        if 'other' in data[column].unique():
            data = data.drop([column + '_other'], axis=1)
        elif 'missing' in data[column].unique():
            data = data.drop([column + '_missing'], axis=1) 
        else:
            excl_val = list(data[column].sort_values(na_position='last', ascending=False).unique())[0]
            #print(excl_val[-1])
            data = data.drop([column + '_{}'.format(excl_val)], axis=1)        

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, [0])
    ])

data_ohe = preprocessor.fit_transform(data)
'''

def create_dummies(data, features):
    """One-hot encode categorical data"""
    #categoricals = list(data.select_dtypes(include='object').columns)
    for column in features:
        if 'other' in data[column].unique():
            data = pd.get_dummies(data, columns=[column], prefix=[column],
                                  drop_first=False)
            data = data.drop([column + '_other'], axis=1)     
        else:
            data = pd.get_dummies(data, columns=[column], prefix=[column],
                                  drop_first=True) 
    return data

def impute_mean(data, features):
    """Fill continuous variables with mean value"""
    for column in features:
        data["{}_i".format(column)] = data[column].replace(np.nan, data[column].mean())
        data.drop(columns=[column], axis=1, inplace=True)
    return data

def impute_cat(data, features):
    """Fill categorical variables with max value"""
    for column in features:
        data["{}_i".format(column)] = data[column].fillna(data[column].value_counts().idxmax())
        data.drop(columns=[column], axis=1, inplace=True)
    return data

def show_bins(bins):
    """Shows the ranges of values the bins comprises"""
    
    for i in range(len(bins)-1):
        print(f'Bin {i}: {str(bins[i])} -> {str(bins[i+1])}')

def descretization(data, features, bins=10, strategy='quantile'):
    for column in features:
        kbins = KBinsDiscretizer(n_bins=bins, strategy=strategy, encode='ordinal')
        data["{}_d".format(column)] = kbins.fit_transform(np.array(data[column]).reshape(-1,1))
        print("Bins:", column)
        print(show_bins(kbins.bin_edges_[0]))
        data.drop(columns=[column], axis=1, inplace=True)
    return data

def log_transform(data, features):
    for column in features:
        data["l{}".format(column)] = np.log([x+1 for x in data[column]])
        data.drop(columns=[column], axis=1, inplace=True)
    return data



##SETUP##
sns.set()
random.seed(1234)
plt.rcParams.update(plt.rcParamsDefault)
plt.rcParams.update({'figure.figsize': (12, 10)})
pd.set_option("display.max_columns",None)
pd.set_option("display.max_colwidth", None)
#np.set_printoptions(threshold=sys.maxsize)
np.set_printoptions(threshold=0)
##       to get rid of warning messages
warnings.filterwarnings('ignore') 
##        Remove scientific notations and display numbers with 2 decimal points instead        
pd.options.display.float_format = '{:,.2f}'.format
##        Update default background style of plots        
#sns.set_style(style=darkgrid)

######################################################################################### 

##PATHS##
pth=r'C:\Users\SALDRU1\Dropbox\PC\Desktop\ml_lres\lab_wb_fin_incl\\'

##CANVAS VARIABLES##
train = "training.csv"
test = "test.csv"
```

1. Extracting region data based on approximate longitude, latitude coordinates using argis reverse geocoding method.

2. Computing the number of financial institutions or access points within a region.

Given that most of our variables are binary or categorical, we use cramers v statistic to measure association between variables. 

```python
#Alternative for categorical/nominal values
#In statistics, Cramr's V (sometimes referred to as Cramr's phi and denoted as c) is a measure 
#of association between two nominal variables, giving a value between 0 and +1 (inclusive). 
#It is based on Pearson's chi-squared statistic and was published by Harald Cramr in 1946.[1]
#Cramr's V varies from 0 (corresponding to no association between the variables) to 1 
#(complete association) and can reach 1 only when each variable is completely determined by the other.

# Using association_metrics library
import association_metrics as am
# exclue continuous columns
cat_cols = df_train0.select_dtypes(exclude='float').columns.difference(['train'])
#cat_cols = sorted(list(set(df_train0.select_dtypes(exclude='float').columns)-set(['train'])))
# Convert object columns to Category columns
df = df_train0[cat_cols].apply(lambda x: x.astype("category"))
# Initialize a CramersV object using the pandas.DataFrame (df)
cramers_v = am.CramersV(df)
# It will return a pairwise matrix filled with Cramer's V, where 
# columns and index are the categorical variables of the passed     # pandas.DataFrame
cfit = cramers_v.fit().round(2)

sns.heatmap(cfit,
            square=True, cmap='RdBu_r', linewidths=.5,
            annot=False, fmt='.2f').figure.tight_layout()

plt.show()
```
