Based on the count of each section, we can calculate precision and recall of each label:

*   **Precision** is a measure of the accuracy provided that a class label has been predicted. It is defined by: precision = TP/(TP+FP) - i.e. how many were positive predictions did we get correct given the number of positive predictions we made 

*   **Recall** is the true positive rate. It is defined as: Recall = TP/(TP+FN) - i.e. how many positive predictions did we get correct given that the real outcome is positive.

So, we can calculate the precision and recall of each class.

**F1 score:**
Now we are in the position to calculate the F1 scores for each label based on the precision and recall of that label. It is defined by: F1 score = 2 * (prc * rec)/(prc + rec)

The F1 score is the harmonic average of theprecision and recall, where an F1score reaches its best value at 1 (perfect precision and recall) and worst at 0. It is a good way to show that a classifer has a good value for both recall and precision.

Finally, we can tell the average accuracy for this classifier is the average of the F1-score for both labels, which is 0.62 in our case.

- ROC curve (specificity vs sensitivity)
Moving forward, we may want to find some combination of specificity and sensitivity. Its ROC analysis where the tradeoff between specificity and sensitivity is explored as a trade off between TPR and FPR (that is, recall and fallout).

Focusing only on sensitivity (probability of positive test given that an individual opens a mobile money account) means we minimize false negatives. This means there will be few people we incorrectly predict to have an account, but more people we incorrectly predict to not have an account. A negative result in a test with high sensitivity is useful for ruling out account holders.
Focusing only on specificity (probability of negative test given that individual does not open a mobile money account) means we minimize false positives. This means there will be few people we incorrectly predict to not have an account, but more people we incorrectly predict to have an account. A positive result in a test with high specificity is useful for ruling in account holders.
